```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pickle as pkl
```

```python
with open("experimento.pkl", "rb") as f:
    data = pkl.load(f)
```

```python
dict = {}
for k, v in data.items():
    dict[k] = [v[i][0] for i in range(len(v))]
```

```python
df = pd.DataFrame.from_dict(dict)
```

```python
df.describe()
```

```python
df.melt()
```

```python
plt.rcParams.update({'font.size': 20})
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot()
l = ["HCI - Função 1 - Restrição 1", "HCI - Função 1 - Restrição 2", "HCI - Função 2 - Restrição 1", 
          "HCI - Função 2 - Restrição 2", "BVNS - Função 1 - Restrição 1", "BVNS - Função 1 - Restrição 2", 
          "BVNS - Função 2 - Restrição 1", "BVNS - Função 2 - Restrição 2"]
g = sns.boxplot(y = "value", x="variable", data=df.melt(), ax=ax)
ax.set_xticklabels(labels=l, rotation=45)
plt.xlabel("", fontweight="bold")
plt.ylabel("Valor Mínimo", fontweight="bold")
plt.title("Variação das funções objetivo", fontweight="bold")
plt.savefig('otimizações.png', format='png',dpi = 300, bbox_inches='tight',)
```
