```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pickle as pkl
```

```python
with open("experimento.pkl", "rb") as f:
    data = pkl.load(f)
```

```python
dict = {}
for k, v in data.items():
    dict[k] = [v[i][0] for i in range(len(v))]
```

```python
df = pd.DataFrame.from_dict(dict)
```

```python
df.describe()
```

```python
df.melt()
```
